```python
import os
from langchain.document_loaders import TextLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.embeddings import OpenAIEmbeddings
from langchain.vectorstores import Chroma
```

```python
# Setting the environment path and autheticating the OpenAI API

os.environ["OPENAI_API_KEY"] = "Your API Key"
```

```python
#Need to load the document

loader = TextLoader("/Users/manuvenugopal/Downloads/LangChainNotebooks-UNZIP-ME/01-Data-Connections/some_data/FDR_State_of_Union_1944.txt")
data = loader.load()

print(data)
```

```python
#Need to split the data
text_splitter = CharacterTextSplitter(chunk_size = 500)
docs = text_splitter.split_documents(data)

docs
```

```python
pwd
```

```python
# Need to embedd the above split documents
# Need to stores the vector to chroma db

embedding_function = OpenAIEmbeddings()
db = Chroma.from_documents(docs, embedding_function, persist_directory = "./speech_new_db")
db.persist()
```

```python
# Suppose we need to find the closest similarity match from the stored vactors

query = "What did FDR said about food costs?"

new_db = Chroma(persist_directory="./speech_new_db", embedding_function=embedding_function)
similar_docs = new_db.similarity_search(query)

similar_docs
```
